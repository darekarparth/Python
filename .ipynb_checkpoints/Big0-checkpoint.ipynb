```python
# O(n): Number of times For Loop is running linearly
# O(n): Linear Time. For loop is running twice which becomes o(2n) 
# • Directly proportional to the data set size. 
# • Example: Looping through an array. 
def print_List(n):
    for i in range(n):
         print (i)

    for j in range(n):
         print (j)

print_List(10)
```

```python
# o(n^2): Here we are having For Loop inside another For Loop. Number of operations increased raised to 2
# This type of code is less efficient from time complexity standpoint. 
def print_num(n):
    for i in range(n):
        for j in range(n):
            print(i,j)

print_num(10)
```

```python
# Drop Non-Dominants
def print_num(n):
    # o(n^2)
    for i in range(n):
        for j in range(n):
            print(i,j)
    # o(n)
    for k in range(n):
         print (k)

print_num(10)

# Output : o(n^2) + o(n) = o(n^2 + n). here n^2 is dominant and n is non dominant, hence we drop non dominant term and time complexitiy of above
#  code is o(n^2) 
```

```python
# o(1):
# here the number is increasing constantly. Number of operations will remain constant. THis is very efficient big o
# def add_num(n):
#     return n + n + n

def add_num(n):
    return n + n + n

print(add_num(10))
```

```python
# Big O: Different terms and inputs
def print_List(a,b):
    for i in range(a):
         print (i)

    for j in range(b):
         print (j)

print_List(5,6)
# this will become O(a+b)
```

```python
# Big O: Different terms and inputs
def print_num(a,b):
    for i in range(a):
        for j in range(b):
            print(i,j)

print_num(8,9)
# this will become O(a*b)
```

```python
 # Big O: Lists
# In list anything added or deleted at end has Time Complexity O(1) and anything added or deleted in between or Starting has Time Complexity O(n)
myList = [11,12,43,6]
myList.append(5)  #Time Complexity O(1)
print(myList)

myList.pop(4) #Time Complexity O(1)
print(myList)

myList.insert(3,45) #Time Complexity O(n)
print(myList)

myList.insert(1,8) #Time Complexity O(n)
print(myList)
```
